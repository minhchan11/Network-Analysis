```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from pylab import rcParams
import seaborn as sb

import networkx as nx
%matplotlib inline
rcParams['figure.figsize'] = 5, 4
sb.set_style('whitegrid')
```

# Creating Graph Objects

```python
G = nx.Graph()
nx.draw_networkx(G)
```

```python
G.add_node(1)
nx.draw_networkx(G)
```

```python
G.add_nodes_from([2,3,4,5,6,8,9,12,15,16])
nx.draw_networkx(G)
```

```python
G.add_edges_from([(2,4),(2,6),(2,8),(2,12),(2,16),(3,6),(3,9), (3,12),(3,15),(4,8),(4,12),(4,16),(6,12),(8,16)])
nx.draw_networkx(G)
```

# Different styles of visualization

```python
nx.draw_networkx(G, pos = nx.circular_layout(G))
```

```python
nx.draw_networkx(G, pos = nx.spring_layout(G))
```

# Labeling and coloring

```python
nx.draw_networkx(G, pos = nx.circular_layout(G),node_color='bisque', with_labels=True)
```

```python
G.remove_node(1)
nx.draw_networkx(G, pos = nx.circular_layout(G),node_color='bisque', with_labels=True)
```

# Graph infos

```python
sum_stats = nx.info(G)
print (sum_stats)
```

```python
print (nx.degree(G))
```

# Create graph template from graph generator

```python
G = nx.complete_graph(25)
nx.draw_networkx(G,node_color='bisque', with_labels=True)
```

```python
G = nx.gnc_graph(7, seed=25)
nx.draw_networkx(G,node_color='bisque', with_labels=True)
```
